```python
# Loading Libraries
import pandas as pd
import plotly_express as px
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.rcParams['figure.figsize']=(10,8)
```

## School Id - Analysis

```python
# Loading Data
df = pd.read_csv('../../data_v2/user_educations.csv')
df.shape, df.head()
```

```python
# finding datatypes
df.dtypes
```

```python
# Fixing type variables
type_correction = {'user_id': 'object',
                  'school_id': 'object',
                  'degree_id': 'object'}
df = df.astype(type_correction)
```

```python
# Missing Value Proportions
np.round(df.isnull().sum()/df.shape[0] * 100, 2)
```

```python
# Count of Uniuqe Categories
df.nunique()
```

```python
# Univariate Analysis of school_id
temp = df.school_id.value_counts()
temp.head()
```

- school_id with `999999` are considered to be "others" where the frequency count is marginally lower; fact from Problem Statement.

```python
# Statistics of #peers without `others` group
temp_ = temp.drop(index=999999)
temp_.describe()
```

```python
# Distribution of #peers 
temp_ = pd.DataFrame(temp.drop(index=999999)).reset_index(). \
            rename(columns={'index': 'school_id', 'school_id': '#peers' })
px.box(data_frame=temp_, x='#peers',
       orientation='h', 
       hover_data=['school_id',])
```

```python
# What is number of school having minimum peers signed up ?
len(temp_[temp_['#peers'] == 5])
```
