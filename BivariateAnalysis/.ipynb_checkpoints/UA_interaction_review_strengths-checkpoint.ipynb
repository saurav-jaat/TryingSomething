```python
# Loading Libraries
import pandas as pd
import plotly_express as px
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate

%matplotlib inline
plt.rcParams['figure.figsize']=(10,8)
```

```python
df = pd.read_csv('../../data_v2/interaction_review_strengths.csv')
print(df.shape)
df.head()
```

```python
df.dtypes
```

- `from-to` need to be broken to `reviewer` and `reviewed`

```python
# Breaking from-to columns
from_to = df['from-to'].apply(lambda x: x.split('-'))
from_to = np.array(from_to.to_list())
df['reviewer'] = from_to[:, 0]
df['reviewed'] = from_to[:, 1]
df.head()
```

```python
# Grouping by reviewed_ids
reviewed = pd.pivot_table(df,
              values='reviewer',
              index=['reviewed'],
                columns='strength_id',
              aggfunc='count')
reviewed.head()
```

```python
# Immuting missing values
reviewed.fillna(0, inplace=True)
reviewed.head()
```

```python
reviewed = reviewed.add_prefix("reviewed_strength_").reset_index().rename(columns={'reviewed':'user_id'})
reviewed.head()
```

```python
reviewed = reviewed.astype({'user_id': 'int64'})
reviewed.dtypes
```

- ignore the index_name of `strenght_id`

- lets us see how these values are correlated to the `users_strengths.csv`

```python
# Loading Strenghts
user_strengths = pd.read_csv('../../data_v2/user_strengths.csv')
user_strengths.head()
```

```python
df_merged = pd.merge(user_strengths, reviewed, how='inner', on='user_id')
```

```python
df_merged.head()
```

```python
# Visual plot of relation between strength_1
plt.scatter(df_merged.strength_id_1, df_merged.reviewed_strength_1)
plt.xlabel('Strengths from user_strengths.csv')
plt.ylabel('Count of people who reviewed the same')
plt.title('Strength_id_1')
```

```python
print('Correlation of Strength of Users with with the count of people who reviewed it')
data = [('strength_id', 'correlation'),]
for i in range(1, 9):
    data.append((f'{i}', \
        df_merged['strength_id_'+str(i)].corr(df_merged['reviewed_strength_'+str(i)])
         ))
print(tabulate(data, headers='firstrow'))
```

# Insights
- The correlation between the count of people reviewed a particular `user_id`'s specific `strength_id` and the same from its `user_strengths` data are postively strong.
